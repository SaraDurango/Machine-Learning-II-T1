**1. Simulate any random rectangular matrix A**

```python
import numpy as np

# Define random dimensions for the rectangular matrix: generate a random number between 2 and 10 for the rows and the columns
rows = np.random.randint(2, 10)
columns = np.random.randint(2, 10)

# Ensure that the number of rows and columns is different
while rows == columns:
    columns = np.random.randint(2, 10)

# Generate a random rectangular matrix using the random function
A = np.random.rand(rows, columns)

print("Matrix A:")
print(A)
print()
```

 - ¿What is the rank and trace of A?

```python
# Calculate the rank of A
rank_A = np.linalg.matrix_rank(A)
print("The rank of A is:", rank_A)

# Calculate the trace of A
trace_A = np.trace(A)
print("The trace of A is:", trace_A)
```

- ¿What is the determinant of A?

The determinant of a matrix is defined only for square matrices, and we are working with a rectangular matrix. Therefore, we must first transform it.

To do this, we will find the transpose of A and use the following expression to transform it: A^T * A

```python
# Transform A into a square matrix to find its determinant
A_TA = np.dot(A.T, A)

# Calculate the determinant of A^T * A
determinant_A_TA = np.linalg.det(A_TA)

print("The determinant of A^T * A is:", determinant_A_TA)
```

- ¿Can you invert A? How?

Now that the matrix is square and we know its determinant, we can determine if it is invertible and calculate its inverse.

If the determinant is non-zero, we can calculate its inverse; otherwise, the matrix is not invertible.

```python
# Inverse of A (if possible)
try:
    A_inverse = np.linalg.inv(A_TA)
    print("The inverse of A is:")
    print(A_inverse)
except np.linalg.LinAlgError:
    print("A is not invertible.")
```

- ¿How are eigenvalues and eigenvectors of A’A and AA’ related? What interesting differences can you notice between both?

```python
# Eigenvalues and eigenvectors of A_TA
eigenvalues_AA_T, eigenvectors_AA_T = np.linalg.eig(np.dot(A.T, A))

# Eigenvalues and eigenvectors of A_TA
eigenvalues_A_T_A, eigenvectors_A_T_A = np.linalg.eig(np.dot(A, A.T))

print()
print("Eigenvalues and eigenvectors of A'A:")
print("Eigenvalues:", eigenvalues_AA_T)
print("Eigenvectors:")
print(eigenvectors_AA_T)
print()

print("Eigenvalues and eigenvectors of AA':")
print("Eigenvalues:", eigenvalues_A_T_A)
print("Eigenvectors:")
print(eigenvectors_A_T_A)
```

The difference between the eigenvalues of AA' and A'A lies in the dimensions of the resulting matrices.

When A is an mxn matrix, AA' yields an mxm matrix, whereas A'A yields an nxn matrix.

Similarly, regarding eigenvectors, if A is an mxn matrix, AA' results in an mxm matrix, while A'A results in an nxn matrix.

**Create a PCA object with 2 components:**

```python
# Get Mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# build masks
masktrain = (y_train == 0) | (y_train == 8)
masktest = (y_test == 0) | (y_test == 8)
# Apply masks
X_train = X_train[masktrain]/255.0
y_train = y_train[masktrain]
X_test = X_test[masktest]/255.0
y_test = y_test[masktest]


X_train = X_train.reshape(X_train.shape[0] , X_train.shape[1]* X_train.shape[2])
X_test = X_test.reshape(X_test.shape[0] , X_test.shape[1]* X_test.shape[2])
```

```python
plt.scatter(reduced_x_train[:,0],reduced_x_train[:,1],c=y_train)
plt.show()
```

```python
# pca.fit(X_test)
reduced_x_test = pca.transform(X_test)
print(reduced_x_test.shape)
reduced_x_test = np.real(reduced_x_test)
```

```python
#build the logistic regression
regressor = LogisticRegression(random_state=0)
regressor.fit(reduced_x_train, y_train)
y_pred = regressor.predict(reduced_x_test)
```

```python
#Evaluate whit not numbers to 8
cm = confusion_matrix(y_test, y_pred)
print(cm)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

**9. What are the underlying mathematical principles behind UMAP? What is it useful for?**

UMAP, or Uniform Manifold Approximation and Projection, is a technique for reducing the dimensionality of data. It combines mathematical principles from various fields like nerve theorem, graph theory, Riemannian geometry, and topological data analysis.

UMAP creates a lower-dimensional representation of high-dimensional data while preserving its topological structure. Compared to other methods like T-SNE, UMAP is faster and requires less computational resources. Another advantage of UMAP is that it doesn't need to be recalibrated every time new data points are added.

Here's how UMAP works:

Nerve Theorem: UMAP utilizes the Nerve Theorem to create a simplicial complex that captures the topological properties of the dataset. It constructs a weighted k-nearest neighbor graph where each data point is linked to its k closest neighbors. The weights assigned to these edges reflect the similarity between the connected points.

Riemannian Geometry: In the context of UMAP, Riemannian Geometry is applied to compute distances between points in the lower-dimensional space. This is achieved by minimizing a cost function based on a Riemannian metric, which characterizes the curvature of the space.

Topological Data Analysis (TDA): UMAP leverages TDA, a mathematical discipline focused on the study of data's topological structure. Specifically, TDA aids in constructing a simplicial complex using the Mapper algorithm. This algorithm generates a simplicial complex by considering overlapping sets, allowing for a deeper understanding of the data's topological characteristics.






**10. What are the underlying mathematical principles behind LDA? What is it useful for?**


LDA, or Latent Dirichlet Allocation, is a powerful probabilistic model commonly applied in machine learning and natural language processing to unravel hidden themes within extensive collections of textual data.

LDA operates under the premise that each document comprises a blend of multiple topics, and every word within the document stems from one of these topics. Employing principles from probability theory and topic modeling, the model employs statistical methods to determine the most probable topic distribution for each document and the most likely word distribution for each topic.

This model adopts an unsupervised learning approach, eliminating the need for pre-existing labels or prior knowledge about the dataset. Instead, it autonomously deduces the underlying patterns inherent in the data.

LDA finds diverse applications across various domains, including automated document categorization, news clustering, and the identification of emerging themes within social media discussions.
